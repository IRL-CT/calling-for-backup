```python
import pandas as pd
import numpy as np
import pingouin as pg

# Method 1: Using pingouin (recommended - install with: pip install pingouin)
def calculate_cronbach_alpha_pingouin(data):
    """
    Calculate Cronbach's alpha using pingouin
    
    Parameters:
    data: DataFrame with columns for each questionnaire item
    """
    alpha = pg.cronbach_alpha(data=data)
    return alpha

# Method 2: Manual calculation
def calculate_cronbach_alpha_manual(data):
    """
    Manual calculation of Cronbach's alpha
    
    Parameters:
    data: DataFrame with columns for each questionnaire item
    """
    # Number of items
    k = data.shape[1]
    
    # Variance of individual items
    item_variances = data.var(axis=0, ddof=1)
    
    # Total score for each participant
    total_scores = data.sum(axis=1)
    
    # Variance of total scores
    total_variance = total_scores.var(ddof=1)
    
    # Cronbach's alpha formula
    alpha = (k / (k - 1)) * (1 - (item_variances.sum() / total_variance))
    
    return alpha


# If your columns are named differently, select the questionnaire columns
questionnaire_columns = [
    'Q968',  # Question 1
    'Q969',  # Question 2
    'Q970',  # Question 3
    'Q971',  # Question 4
    'Q972'   # Question 5
]

# Select only the questionnaire items
questionnaire_data = answers[questionnaire_columns]
print(questionnaire_data.head())
#turn into numeric columns
questionnaire_data = questionnaire_data.apply(pd.to_numeric, errors='coerce')

# Remove any rows with missing values
questionnaire_data = questionnaire_data.dropna()

# Calculate Cronbach's alpha
alpha_pingouin = calculate_cronbach_alpha_pingouin(questionnaire_data)
alpha_manual = calculate_cronbach_alpha_manual(questionnaire_data)

print(f"Cronbach's alpha (pingouin): {alpha_pingouin[0]:.3f}")
print(f"Cronbach's alpha (manual): {alpha_manual:.3f}")

# Additional statistics
print(f"\nNumber of items: {questionnaire_data.shape[1]}")
print(f"Number of participants: {questionnaire_data.shape[0]}")
print(f"Mean scores per item:")
print(questionnaire_data.mean().round(2))
print(f"\nStandard deviations per item:")
print(questionnaire_data.std().round(2))

# Interpretation guide
def interpret_alpha(alpha):
    if alpha >= 0.9:
        return "Excellent"
    elif alpha >= 0.8:
        return "Good" 
    elif alpha >= 0.7:
        return "Acceptable"
    elif alpha >= 0.6:
        return "Questionable"
    else:
        return "Poor"

interpretation = interpret_alpha(alpha_manual)
print(f"\nInterpretation: {interpretation}")



#now for the post questionnaire questions
questionnaire_columns_post = [
    'Q979', # Question 1
    'Q980', # Question 2
    'Q981', # Question 3
    'Q982', # Question 4
    'Q983.1' # Question 5
]

# Select only the questionnaire items
questionnaire_data_post = answers[questionnaire_columns_post]
print(questionnaire_data_post.head())
#turn into numeric columns
questionnaire_data_post = questionnaire_data_post.apply(pd.to_numeric, errors='coerce')
# Remove any rows with missing values
questionnaire_data_post = questionnaire_data_post.dropna()
# Calculate Cronbach's alpha
alpha_pingouin_post = calculate_cronbach_alpha_pingouin(questionnaire_data_post)
alpha_manual_post = calculate_cronbach_alpha_manual(questionnaire_data_post)
print(f"Cronbach's alpha post (pingouin): {alpha_pingouin_post[0]:.3f}")
print(f"Cronbach's alpha post (manual): {alpha_manual_post:.3f}")


pre_data = answers[questionnaire_columns]
post_data = answers[questionnaire_columns_post]
#turn into numeric columns
pre_data = pre_data.apply(pd.to_numeric, errors='coerce')
post_data = post_data.apply(pd.to_numeric, errors='coerce')

# Test-retest reliability (if appropriate for your design)
from scipy.stats import pearsonr

# Calculate correlations between pre and post for each item
for ind, col in enumerate(questionnaire_columns):
    r, p = pearsonr(pre_data[col], post_data[questionnaire_columns_post[ind]])
    print(f"{col} test-retest reliability: r = {r:.3f}, p = {p:.3f}")
```

```python
import scipy.stats as stats
import pandas as pd
import numpy as np
from scipy.stats import shapiro, levene
from statsmodels.stats.multitest import multipletests

# Independent t-test comparing difference scores between conditions
def run_independent_ttests(answers_df):
    """
    Compare difference scores between error and control conditions
    """
    
    # Map difference columns to question names
    diff_columns = ['diff 1', 'diff 2', 'diff 3', 'diff 4', 'diff 5']
    question_names = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5']
    
    results = []
    p_values = []
    
    for i, (diff_col, q_name) in enumerate(zip(diff_columns, question_names)):
        # Get difference scores by condition
        error_diffs = answers_df[answers_df['condition'] == 'error'][diff_col].astype(float)
        control_diffs = answers_df[answers_df['condition'] == 'control'][diff_col].astype(float)
        
        # Remove any NaN values
        error_diffs = error_diffs.dropna()
        control_diffs = control_diffs.dropna()
        
        # Descriptive statistics
        error_mean = error_diffs.mean()
        control_mean = control_diffs.mean()
        error_std = error_diffs.std()
        control_std = control_diffs.std()
        
        # Check normality (optional)
        _, p_norm_error = shapiro(error_diffs) if len(error_diffs) > 3 else (None, 0.05)
        _, p_norm_control = shapiro(control_diffs) if len(control_diffs) > 3 else (None, 0.05)
        
        # Check equal variances
        _, p_levene = levene(error_diffs, control_diffs)
        
        # Perform appropriate test
        if p_norm_error > 0.05 and p_norm_control > 0.05:
            # Data is normal, use t-test
            if p_levene > 0.05:
                # Equal variances
                t_stat, p_val = stats.ttest_ind(error_diffs, control_diffs, equal_var=True)
                test_type = "Independent t-test"
            else:
                # Unequal variances (Welch's t-test)
                t_stat, p_val = stats.ttest_ind(error_diffs, control_diffs, equal_var=False)
                test_type = "Welch's t-test"
        else:
            # Data is not normal, use Mann-Whitney U
            t_stat, p_val = stats.mannwhitneyu(error_diffs, control_diffs, alternative='two-sided')
            test_type = "Mann-Whitney U"
        
        # Calculate Cohen's d (effect size)
        pooled_std = np.sqrt(((len(error_diffs)-1)*error_diffs.var() + 
                             (len(control_diffs)-1)*control_diffs.var()) / 
                            (len(error_diffs) + len(control_diffs) - 2))
        
        if pooled_std != 0:
            cohens_d = (error_mean - control_mean) / pooled_std
        else:
            cohens_d = 0
            
        # Store results
        results.append({
            'Question': q_name,
            'Test_Type': test_type,
            'Error_Mean_Diff': error_mean,
            'Control_Mean_Diff': control_mean,
            'Error_SD': error_std,
            'Control_SD': control_std,
            'Test_Statistic': t_stat,
            'p_value': p_val,
            'Cohens_d': cohens_d,
            'n_error': len(error_diffs),
            'n_control': len(control_diffs)
        })
        
        p_values.append(p_val)
    
    # Multiple comparison correction (Bonferroni)
    _, p_corrected, _, _ = multipletests(p_values, alpha=0.01, method='bonferroni')

    print(p_corrected)
    
    # Add corrected p-values to results
    for i, result in enumerate(results):
        result['p_corrected'] = p_corrected[i]
        result['significant'] = p_corrected[i] < 0.05
    
    return pd.DataFrame(results)

# Run the analysis
results_df = run_independent_ttests(answers)
print("=== INDEPENDENT T-TEST RESULTS ===")
print(results_df.round(4))

# Print significant results
significant_results = results_df[results_df['significant']]
if len(significant_results) > 0:
    print("\n=== SIGNIFICANT DIFFERENCES ===")
    for _, row in significant_results.iterrows():
        print(f"{row['Question']}: p = {row['p_corrected']:.4f}, Cohen's d = {row['Cohens_d']:.3f}")
else:
    print("\nNo significant differences found after correction.")
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

def run_mixed_effects_models(answers_df):
    """
    Run mixed-effects models for each question
    """
    
    # Reshape data to long format for mixed models
    # Create pre and post columns properly
    pre_cols = ['Q968', 'Q969', 'Q970', 'Q971', 'Q972']
    post_cols = ['Q979', 'Q980', 'Q981', 'Q982', 'Q983.1']
    question_names = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5']
    
    results_mixed = []
    
    for i, (pre_col, post_col, q_name) in enumerate(zip(pre_cols, post_cols, question_names)):
        # Create long format data for this question
        df_long = []
        
        for _, row in answers_df.iterrows():
            # Pre-interaction
            df_long.append({
                'participant': row['participant'],
                'condition': row['condition'],
                'timepoint': 'pre',
                'response': float(row[pre_col])
            })
            
            # Post-interaction  
            df_long.append({
                'participant': row['participant'],
                'condition': row['condition'],
                'timepoint': 'post',
                'response': float(row[post_col])
            })
        
        df_question = pd.DataFrame(df_long)
        
        # Remove any NaN values
        df_question = df_question.dropna()
        
        # Ensure participant is treated as categorical
        df_question['participant'] = df_question['participant'].astype(str)
        
        # Create dummy variables for the interaction
        df_question['condition_error'] = (df_question['condition'] == 'error').astype(int)
        df_question['timepoint_post'] = (df_question['timepoint'] == 'post').astype(int)
        df_question['interaction'] = df_question['condition_error'] * df_question['timepoint_post']
        
        try:
            # Fit mixed-effects model
            # Random intercept for participants
            model = smf.mixedlm("response ~ condition_error * timepoint_post", 
                               df_question, 
                               groups=df_question["participant"])
            result = model.fit()
            
            # Extract key statistics
            # The interaction term tells us if the change differs between conditions
            interaction_coef = result.params.get('condition_error:timepoint_post', np.nan)
            interaction_pval = result.pvalues.get('condition_error:timepoint_post', np.nan)
            
            # Main effects
            condition_coef = result.params.get('condition_error', np.nan)
            condition_pval = result.pvalues.get('condition_error', np.nan)
            
            timepoint_coef = result.params.get('timepoint_post', np.nan)
            timepoint_pval = result.pvalues.get('timepoint_post', np.nan)
            
            results_mixed.append({
                'Question': q_name,
                'Condition_Effect': condition_coef,
                'Condition_p': condition_pval,
                'Timepoint_Effect': timepoint_coef,
                'Timepoint_p': timepoint_pval,
                'Interaction_Effect': interaction_coef,
                'Interaction_p': interaction_pval,
                'AIC': result.aic,
                'n_observations': len(df_question)
            })
            
            print(f"\n=== {q_name} Mixed-Effects Model ===")
            print(f"Interaction effect (condition Ã— time): Î² = {interaction_coef:.4f}, p = {interaction_pval:.4f}")
            
        except Exception as e:
            print(f"Error fitting model for {q_name}: {e}")
            results_mixed.append({
                'Question': q_name,
                'Condition_Effect': np.nan,
                'Condition_p': np.nan,
                'Timepoint_Effect': np.nan,
                'Timepoint_p': np.nan,
                'Interaction_Effect': np.nan,
                'Interaction_p': np.nan,
                'AIC': np.nan,
                'n_observations': len(df_question)
            })
    
    # Multiple comparison correction for interaction effects
    interaction_pvals = [r['Interaction_p'] for r in results_mixed if not np.isnan(r['Interaction_p'])]
    if len(interaction_pvals) > 0:
        _, p_corrected_interaction, _, _ = multipletests(interaction_pvals, alpha=0.01, method='bonferroni')
        
        # Add corrected p-values
        j = 0
        for i, result in enumerate(results_mixed):
            if not np.isnan(result['Interaction_p']):
                result['Interaction_p_corrected'] = p_corrected_interaction[j]
                result['Interaction_significant'] = p_corrected_interaction[j] < 0.05
                j += 1
            else:
                result['Interaction_p_corrected'] = np.nan
                result['Interaction_significant'] = False
    
    return pd.DataFrame(results_mixed)

# Run mixed-effects analysis
mixed_results = run_mixed_effects_models(answers)
print("\n=== MIXED-EFFECTS MODEL RESULTS ===")
print(mixed_results.round(4))
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
from scipy.stats import shapiro, wilcoxon, ttest_rel
from statsmodels.stats.multitest import multipletests
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_pre_post_changes(answers_df):
    """
    Analyze pre-post changes across all participants, ignoring conditions
    """
    
    # Define the question pairs (pre and post)
    question_pairs = [
        ('Q968', 'Q979', 'Q1'),
        ('Q969', 'Q980', 'Q2'), 
        ('Q970', 'Q981', 'Q3'),
        ('Q971', 'Q982', 'Q4'),
        ('Q972', 'Q983.1', 'Q5')
    ]
    
    results = []
    p_values = []
    
    print("=== PRE-POST ANALYSIS (ALL PARTICIPANTS) ===\n")
    
    for pre_col, post_col, q_name in question_pairs:
        # Get pre and post scores (convert to numeric)
        pre_scores = pd.to_numeric(answers_df[pre_col], errors='coerce')
        post_scores = pd.to_numeric(answers_df[post_col], errors='coerce')
        
        # Remove pairs with missing data
        valid_mask = ~(pre_scores.isna() | post_scores.isna())
        pre_valid = pre_scores[valid_mask]
        post_valid = post_scores[valid_mask]
        
        # Calculate difference scores
        diff_scores = post_valid - pre_valid
        
        # Descriptive statistics
        pre_mean = pre_valid.mean()
        post_mean = post_valid.mean()
        pre_std = pre_valid.std()
        post_std = post_valid.std()
        mean_diff = diff_scores.mean()
        diff_std = diff_scores.std()
        
        # Test normality of differences
        if len(diff_scores) >= 3:
            _, normality_p = shapiro(diff_scores)
        else:
            normality_p = 0.05  # Assume non-normal if too few observations
            
        # Choose appropriate test
        if normality_p > 0.05:
            # Normal distribution - use paired t-test
            t_stat, p_val = ttest_rel(pre_valid, post_valid)
            test_used = "Paired t-test"
        else:
            # Non-normal distribution - use Wilcoxon signed-rank test
            t_stat, p_val = wilcoxon(pre_valid, post_valid, zero_method='wilcox')
            test_used = "Wilcoxon signed-rank"
        
        # Calculate Cohen's d for effect size
        if diff_std != 0:
            cohens_d = mean_diff / diff_std
        else:
            cohens_d = 0
            
        # Interpret effect size
        if abs(cohens_d) < 0.2:
            effect_size = "Negligible"
        elif abs(cohens_d) < 0.5:
            effect_size = "Small"
        elif abs(cohens_d) < 0.8:
            effect_size = "Medium"
        else:
            effect_size = "Large"
            
        results.append({
            'Question': q_name,
            'n_pairs': len(pre_valid),
            'Pre_Mean': pre_mean,
            'Pre_SD': pre_std,
            'Post_Mean': post_mean,
            'Post_SD': post_std,
            'Mean_Difference': mean_diff,
            'Diff_SD': diff_std,
            'Test_Used': test_used,
            'Test_Statistic': t_stat,
            'p_value': p_val,
            'Cohens_d': cohens_d,
            'Effect_Size': effect_size,
            'Normality_p': normality_p
        })
        
        p_values.append(p_val)
        
        # Print individual results
        print(f"=== {q_name} ===")
        print(f"N pairs: {len(pre_valid)}")
        print(f"Pre:  M = {pre_mean:.3f}, SD = {pre_std:.3f}")
        print(f"Post: M = {post_mean:.3f}, SD = {post_std:.3f}")
        print(f"Difference: M = {mean_diff:.3f}, SD = {diff_std:.3f}")
        print(f"Test: {test_used}")
        print(f"p-value (uncorrected): {p_val:.4f}")
        print(f"Cohen's d: {cohens_d:.3f} ({effect_size})")
        print()
    
    # Apply Bonferroni correction
    _, p_corrected, _, _ = multipletests(p_values, alpha=0.01, method='bonferroni')
    
    # Add corrected p-values to results
    for i, result in enumerate(results):
        result['p_corrected'] = p_corrected[i]
        result['significant'] = p_corrected[i] < 0.05
        
    return pd.DataFrame(results)

def create_pre_post_visualizations(answers_df):
    """
    Create visualizations for pre-post comparisons
    """
    
    # Question pairs
    question_pairs = [
        ('Q968', 'Q979', 'Q1'),
        ('Q969', 'Q980', 'Q2'), 
        ('Q970', 'Q981', 'Q3'),
        ('Q971', 'Q982', 'Q4'),
        ('Q972', 'Q983.1', 'Q5')
    ]
    
    # Create subplot figure
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.flatten()
    
    for i, (pre_col, post_col, q_name) in enumerate(question_pairs):
        ax = axes[i]
        
        # Get valid data
        pre_scores = pd.to_numeric(answers_df[pre_col], errors='coerce')
        post_scores = pd.to_numeric(answers_df[post_col], errors='coerce')
        
        valid_mask = ~(pre_scores.isna() | post_scores.isna())
        pre_valid = pre_scores[valid_mask]
        post_valid = post_scores[valid_mask]
        
        # Create paired plot
        ax.scatter(pre_valid, post_valid, alpha=0.6, s=50)
        
        # Add diagonal line (no change line)
        min_val = min(pre_valid.min(), post_valid.min())
        max_val = max(pre_valid.max(), post_valid.max())
        ax.plot([min_val, max_val], [min_val, max_val], 'r--', alpha=0.5, label='No change')
        
        ax.set_xlabel('Pre-interaction Score')
        ax.set_ylabel('Post-interaction Score')
        ax.set_title(f'{q_name}: Pre vs Post Scores')
        ax.legend()
        ax.grid(True, alpha=0.3)
        
        # Set equal aspect ratio
        ax.set_aspect('equal', adjustable='box')
    
    # Remove the extra subplot
    axes[-1].remove()
    
    plt.tight_layout()
    plt.show()
    
    # Box plot of differences
    fig, ax = plt.subplots(1, 1, figsize=(10, 6))
    
    diff_data = []
    question_names = []
    
    for pre_col, post_col, q_name in question_pairs:
        pre_scores = pd.to_numeric(answers_df[pre_col], errors='coerce')
        post_scores = pd.to_numeric(answers_df[post_col], errors='coerce')
        
        valid_mask = ~(pre_scores.isna() | post_scores.isna())
        differences = post_scores[valid_mask] - pre_scores[valid_mask]
        
        diff_data.extend(differences.tolist())
        question_names.extend([q_name] * len(differences))
    
    diff_df = pd.DataFrame({'Question': question_names, 'Difference': diff_data})
    
    sns.boxplot(data=diff_df, x='Question', y='Difference', ax=ax)
    ax.axhline(y=0, color='red', linestyle='--', alpha=0.7, label='No change')
    ax.set_title('Distribution of Pre-Post Differences by Question')
    ax.set_ylabel('Post - Pre Score')
    ax.legend()
    plt.show()

# Run the analysis
results_df = analyze_pre_post_changes(answers)

# Display results table
print("=== SUMMARY TABLE ===")
display_cols = ['Question', 'n_pairs', 'Pre_Mean', 'Post_Mean', 'Mean_Difference', 
                'Test_Used', 'p_value', 'p_corrected', 'Cohens_d', 'Effect_Size', 'significant']
print(results_df[display_cols].round(4))

# Print significant results
significant_results = results_df[results_df['significant']]
if len(significant_results) > 0:
    print("\n=== SIGNIFICANT CHANGES (after correction) ===")
    for _, row in significant_results.iterrows():
        direction = "increased" if row['Mean_Difference'] > 0 else "decreased"
        print(f"{row['Question']}: Scores {direction} significantly")
        print(f"  Mean change: {row['Mean_Difference']:.3f}")
        print(f"  p-corrected: {row['p_corrected']:.4f}")
        print(f"  Effect size: {row['Cohens_d']:.3f} ({row['Effect_Size']})")
        print()
else:
    print("\n=== NO SIGNIFICANT CHANGES FOUND (after correction) ===")

# Create visualizations
create_pre_post_visualizations(answers)

# Additional summary statistics
print("\n=== OVERALL SUMMARY ===")
print(f"Total number of participant pairs analyzed: {results_df['n_pairs'].iloc[0]}")
print(f"Questions showing significant change: {len(significant_results)}/5")

# Calculate overall pre-post correlation
print("\n=== PRE-POST CORRELATIONS ===")
question_pairs = [
    ('Q968', 'Q979', 'Q1'),
    ('Q969', 'Q980', 'Q2'), 
    ('Q970', 'Q981', 'Q3'),
    ('Q971', 'Q982', 'Q4'),
    ('Q972', 'Q983.1', 'Q5')
]

for pre_col, post_col, q_name in question_pairs:
    pre_scores = pd.to_numeric(answers[pre_col], errors='coerce')
    post_scores = pd.to_numeric(answers[post_col], errors='coerce')
    
    valid_mask = ~(pre_scores.isna() | post_scores.isna())
    if valid_mask.sum() > 2:
        correlation, cor_p = stats.pearsonr(pre_scores[valid_mask], post_scores[valid_mask])
        print(f"{q_name}: r = {correlation:.3f}, p = {cor_p:.4f}")
```
